```python
import math
nums = [1,2,4]

def find_all_subsets(nums):
    if len(nums) == 0:
        return [[]]
    value = nums.pop()
    smaller_subsets = find_all_subsets(nums)
    newset = smaller_subsets.copy()
    for sets in smaller_subsets:
        dset = sets.copy()
        dset.append(value)
        newset.append(dset)
    return newset

print(find_all_subsets(nums))
```
